```python
from sklearn.linear_model import RidgeCV
import numpy as np

X_train = np.load('/sietch_colab/akapoor/XPopVAE/experiments/IM_symmetric/processed_data/0/rep0/discovery_train.npy')  # (4000, 10000)
y_train = np.load('/sietch_colab/akapoor/XPopVAE/phenotype_creation/simulated_phenotype_train.npy')
X_val = np.load('/sietch_colab/akapoor/XPopVAE/experiments/IM_symmetric/processed_data/0/rep0/discovery_val.npy')
y_val = np.load('/sietch_colab/akapoor/XPopVAE/phenotype_creation/simulated_phenotype_val.npy')

ridge = RidgeCV(alphas=np.logspace(-2, 6, 50), cv=5)
ridge.fit(X_train, y_train)
print("Raw genotype train R²:", ridge.score(X_train, y_train))
print("Raw genotype val R²:", ridge.score(X_val, y_val))
```

```python
print("Selected alpha:", ridge.alpha_)
```
